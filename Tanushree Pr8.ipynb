```python
import pandas as pd 
import matplotlib.pyplot as plt 
data = "tipa.csv" 
df = pd.read_csv(data) 
df.head()
```

```python
smoker_counts = df['Smoking'].value_counts() 
labels = ['Non-Smoker', 'Smoker'] 
sizes = [smoker_counts['No'], smoker_counts['Yes']]  
# Assuming 'Yes' and 'No' as values 
colors = ['lightblue', 'salmon']  
plt.figure(figsize=(7, 7)) 
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True) 
plt.title("Percentage of People in Smoking vs Non-Smoking Areas") 
plt.show()
```
