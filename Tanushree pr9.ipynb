```python
import nltk
from nltk.corpus import inaugural  # Example dataset from NLTK
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download necessary NLTK datasets (if not already available)
nltk.download('inaugural')

# Load a sample text (Inaugural addresses from U.S. Presidents)
text = ' '.join(inaugural.words('2009-Obama.txt'))  # Example: Barack Obama's 2009 speech

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Hide axis
plt.title("Word Cloud from Barack Obama's 2009 Inaugural Speech")
plt.show()
```
